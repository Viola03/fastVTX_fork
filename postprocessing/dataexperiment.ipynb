```python
import numpy as np
import matplotlib.pyplot as plt

mother_endvertex_x = file["DAUGHTER1_TRUEENDVERTEX_X"].array(library="np")
mother_endvertex_y = file["DAUGHTER1_TRUEENDVERTEX_Y"].array(library="np")
mother_endvertex_z = file["DAUGHTER1_TRUEENDVERTEX_Z"].array(library="np")

mother_ipchi2 = file["MOTHER_IPCHI2_OWNPV"].array(library="np")
mother_fdchi2 = file["MOTHER_FDCHI2_OWNPV"].array(library="np")
mother_chi2 = file["MOTHER_ENDVERTEX_CHI2"].array(library="np")
```

```python
def plot_histogram(data, bins, range_, xlabel, title, filename):
    plt.hist(data, bins=bins, range=range_, histtype='step', color='blue')
    plt.xlabel(xlabel)
    plt.ylabel("Entries")
    plt.title(title)
    plt.show()
```

```python
mother_ipchi2 = file["MOTHER_IPCHI2_OWNPV"].array()
plot_histogram(mother_ipchi2, bins=50, range_=(0, 20), xlabel="IPChi2 (MOTHER)", title="MOTHER IPChi2", filename="MOTHER_IPCHI2_OWNPV")

daughter1_ipchi2 = file["DAUGHTER1_IPCHI2_OWNPV"].array()
plot_histogram(daughter1_ipchi2, bins=50, range_=(0, 20), xlabel="IPChi2 (DAUGHTER1)", title="DAUGHTER1 IPChi2", filename="DAUGHTER1_IPCHI2_OWNPV")

mother_fdchi2 = file["MOTHER_FDCHI2_OWNPV"].array()
plot_histogram(mother_fdchi2, bins=50, range_=(0, 200), xlabel="FDChi2 (MOTHER)", title="MOTHER FDChi2", filename="MOTHER_FDCHI2_OWNPV")

intermediate_fdchi2 = file["INTERMEDIATE_FDCHI2_OWNPV"].array()
plot_histogram(intermediate_fdchi2, bins=50, range_=(0, 200), xlabel="FDChi2 (INTERMEDIATE)", title="INTERMEDIATE FDChi2", filename="INTERMEDIATE_FDCHI2_OWNPV")

daughter1_chi2ndof = file["DAUGHTER1_TRACK_CHI2NDOF"].array()
plot_histogram(daughter1_chi2ndof, bins=50, range_=(0, 5), xlabel="Chi2/NDOF (DAUGHTER1)", title="DAUGHTER1 Track Fit Chi2/NDOF", filename="DAUGHTER1_TRACK_CHI2NDOF")

```

```python
mother_px = file["MOTHER_PX"].array()
plot_histogram(mother_px, bins=50, range_=(-10e3, 10e3), xlabel="Px (MOTHER)", title="MOTHER Px", filename="MOTHER_PX")

mother_py = file["MOTHER_PY"].array()
plot_histogram(mother_py, bins=50, range_=(-10e3, 10e3), xlabel="Py (MOTHER)", title="MOTHER Py", filename="MOTHER_PY")

mother_pz = file["MOTHER_PZ"].array()
plot_histogram(mother_pz, bins=50, range_=(-10e3, 10e3), xlabel="Pz (MOTHER)", title="MOTHER Pz", filename="MOTHER_PZ")
```

```python
#Renaming branches
import uproot
import awkward as ak

input_file = r"C:\Users\viola\OneDrive\Desktop\PPM12\fast_vertexing_variables-master\inference\example\combinatorial_select_Kuu.root"
tree_name = "DecayTree" 

# Open the ROOT file
with uproot.open(input_file) as file:
    tree = file[tree_name]

    # Load all branches into a dictionary of arrays
    data = tree.arrays()

# Define a dictionary of old branch names to new branch names
branch_rename_map = {
    "MOTHER_TRUEP_X": "MOTHER_PX_TRUE",  
    "MOTHER_TRUEP_Y": "MOTHER_PY_TRUE",
    "MOTHER_TRUEP_Z": "MOTHER_PZ_TRUE",
    "MOTHER_P": "MOTHER_P_TRUE",
    "MOTHER_PT": "MOTHER_PT_TRUE",
    
}

# renamed_data = {}
# for old_name in data.keys():
#     array = data[old_name]  # Access the array using the old name
#     if old_name in branch_rename_map:
#         new_name = branch_rename_map[old_name]
#         renamed_data[new_name] = array  # Store under the new name
#     else:
#         renamed_data[old_name] = array  # Keep the branch as is if no renaming is required

renamed_data = {}
for old_name in ak.fields(data):  # Use ak.fields to get branch names
    array = data[old_name]  # Access the array using the old name
    if old_name in branch_rename_map:
        new_name = branch_rename_map[old_name]
        renamed_data[new_name] = array  # Store under the new name
    else:
        renamed_data[old_name] = array  # Keep the branch as is if no renaming is required



output_file = r"C:\Users\viola\OneDrive\Desktop\PPM12\fast_vertexing_variables-master\inference\example\combinatorial_select_Kuu_renamed.root"
with uproot.recreate(output_file) as new_file:
    new_file[tree_name] = renamed_data

print(f"Branches renamed and written to {output_file}")
```
